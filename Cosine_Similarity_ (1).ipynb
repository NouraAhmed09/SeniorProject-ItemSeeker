```python

import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

nltk.download('stopwords')

# Get the English stop words
stop_words = set(stopwords.words('english'))

# Create a Porter stemmer object
stemmer = PorterStemmer()

def clean_and_stem_text(text):
    # Remove punctuation and convert to lowercase
    text = re.sub(r'[^\w\s]', '', text).lower()

    # Tokenize the text into individual words
    tokens = text.split()

    # Remove stop words and stem the remaining words
    cleaned_tokens = [stemmer.stem(token) for token in tokens if token not in stop_words]

    # Join the cleaned tokens back into a string
    cleaned_text = ' '.join(cleaned_tokens)

    return cleaned_text


# Apply the function to the 'Caption' column
data_caption['Processed_Caption'] = data_caption['Caption'].apply(clean_and_stem_text)
```

```python

# Convert text to lowercase
data_caption['Processed_Caption'] = data_caption['Processed_Caption'].str.lower()

# Remove special characters
data_caption['Processed_Caption'] = data_caption['Processed_Caption'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)

data_caption
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

captions = data_caption['Processed_Caption'].tolist()

vectorizer = TfidfVectorizer()

tfidf_vectors = vectorizer.fit_transform(captions)

print(tfidf_vectors.shape)
```

```python

import cv2
import numpy as np
from PIL import Image

def preprocess_images(image_paths, target_size=(224, 224)):

  processed_images = []
  for image_path in image_paths:
    try:
      img = cv2.imread(image_path)

      img = cv2.resize(img, target_size)

      img_array = np.array(img)

      processed_images.append(img_array)
    except Exception as e:
      print(f"An error occurred while processing the image {image_path}: {e}")
  return processed_images
image_paths = data_caption['Image Name'].tolist()
processed_images = preprocess_images(image_paths)

```

```python
processed_images = np.array(processed_images)
processed_images.shape
```

```python

from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_similarity(text_vector, tfidf_vectors):

  similarity_scores = cosine_similarity(text_vector, tfidf_vectors)
  return similarity_scores[0]

new_text = "black iphone"
new_text_vector = vectorizer.transform([clean_and_stem_text(new_text)])

similarity_scores = calculate_cosine_similarity(new_text_vector, tfidf_vectors)

print(similarity_scores)
```

```python

# get the index of the 3 big value of similarty and plot the image corresponding

import matplotlib.pyplot as plt
import numpy as np
top_3_indices = np.argpartition(similarity_scores, -3)[-3:]

for index in top_3_indices:
  image_path = data_caption['Image Name'][index]
  img = cv2.imread(image_path)
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.title(f"Similarity Score: {similarity_scores[index]:.2f}")
  plt.axis('off')
  plt.show()
```
